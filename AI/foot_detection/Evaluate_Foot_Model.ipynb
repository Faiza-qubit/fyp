```python
from ultralytics import YOLO
import torch

# Load model
model = YOLO("best.pt")

results = model.val(
    data="foot_dataset/data.yaml",         
    imgsz=640,
    batch=8,                                
    plots=True,
    conf=0.25,
    iou=0.6,
    device="cpu"
)
print("\nValidation finished!")
```

```python
def safe_format(value):
    return f"{value:.4f}" if isinstance(value, (int, float)) else "N/A"

print("=== Foot Pose Model Evaluation Report ===")
print(f"Box mAP@0.5:            {results.box.map50:.4f}")
print(f"Box mAP@0.5:0.95:        {results.box.map:.4f}")
print(f"Keypoint mAP@0.5:        {results.pose.map50:.4f}")
print(f"Keypoint mAP@0.5:0.95:   {results.pose.map:.4f}")
print(f"Precision (boxes):       {results.box.p[0]:.4f}")
print(f"Recall (boxes):          {results.box.r[0]:.4f}")
print(f"F1 Score (boxes):        {results.box.f1[0]:.4f}")
print(f"Precision (pose):       {safe_format(results.pose.p[0] if hasattr(results.pose, 'p') and len(results.pose.p) > 0 else None)}")
print(f"Recall (pose):          {safe_format(results.pose.r[0] if hasattr(results.pose, 'r') and len(results.pose.r) > 0 else None)}")
print(f"F1 Score (pose):        {safe_format(results.pose.f1[0] if hasattr(results.pose, 'f1') and len(results.pose.f1) > 0 else None)}")
```

```python
from IPython.display import Image, display, HTML
import os
import glob

results_dir = "results"

if not os.path.exists(results_dir):
    print("Results folder not found.")
else:
    png_paths = sorted(glob.glob(os.path.join(results_dir, "*.png")))

    if not png_paths:
        print("No PNG images found in results/.")
    else:
        html = '<div style="display: flex; flex-wrap: wrap; gap: 20px; justify-content: center;">'

        for path in png_paths:
            fname = os.path.basename(path)
            rel_path = path.replace("\\", "/")  
            html += f'''
            <div style="text-align: center; flex: 1 1 45%; max-width: 48%;">
                <img src="{rel_path}" style="max-width:100%; border: 1px solid #ccc; border-radius: 6px;">
                <div style="margin-top: 8px; font-weight: bold; font-size: 14px;">{fname}</div>
            </div>
            '''
        html += '</div>'
        display(HTML(html))
```
